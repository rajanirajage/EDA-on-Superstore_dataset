```python
# we will perform some kind of visualization with data

df_country=df[['Segment','State','City','Profit']]
df_country.head()
```

```python
df_city=df_country.groupby('City')['Profit'].mean()
```

```python
df_city.sort_values().tail(10).plot(kind='barh',title='Top 10 city in profit')
plt.show()
```

```python
df_city.sort_values().head(10).plot(kind='barh',title='Top 10 city in profit')
plt.show()
# it show lowest profitable city
```

```python
# Average profit with the help of group by
df.groupby('Segment')['Profit'].mean().sort_values().plot(kind='bar',title='Segment wise profit')
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.show()
```

```python
# # see the difference between these two plot and difference between sum and mean
df.groupby('Segment')['Profit'].sum().sort_values().plot(kind='bar',title='Segment wise profit')
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.show()
```
